# **Conditional Statements**

Decision making is required when we want to execute a code only if a certain condition is satisfied.

The if…elif…else statement is used in Python3 for decision making.



---


**Python if Statement Syntax**

if test expression:

    statement(s)
    
![alt text](https://www.tutorialspoint.com/python3/images/if_statement.jpg)
 
    

```
num = 3
if num > 0:
    print(num, "is a positive number.")
print("This is always printed.")

num = -1
if num > 0:
    print(num, "is a positive number.")
print("This is also always printed.")
```

**Python if...else Statement**

***Syntax of if...else***

if test expression:

    Body of if
    
else:

    Body of else
    
   ![alt text](https://www.tutorialspoint.com/python3/images/if_else_statement.jpg)

```
num = 35

if num >= 0:
    print("Positive or Zero")
else:
    print("Negative number")
```

**Python if...elif...else Statement**

***Syntax of if...elif...else***

if expression1:

      statement(s)
   
elif expression2:
   
      statement(s)
      
 elif expression3:
 
      statement(s)
else:

      statement(s)
 
 
 ![alt text](https://tecadmin.net/tutorial/wp-content/uploads/2018/02/if-else-if-statement-flowchart.png)

```
num = int(input("enter number : "))
if num%2 == 0:
   if num%3 == 0:
      print ("Divisible by 3 and 2")
   else:
      print ("divisible by 2 not divisible by 3")
else:
   if num%3 == 0:
      print ("divisible by 3 not divisible by 2")
   else:
      print  ("not Divisible by 2 not divisible by 3")
```
